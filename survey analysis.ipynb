## Importing the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import data from the pre and post survey forms

```python
pre= pd.read_csv("/Users/saitejasriyerramsetti/Library/CloudStorage/GoogleDrive-saitejasriy10@gmail.com/My Drive/Zip LaunchPad/E- Track/Notebooks/Merged_Pre_Survey_Data.csv")
post = pd.read_csv("/Users/saitejasriyerramsetti/Library/CloudStorage/GoogleDrive-saitejasriy10@gmail.com/My Drive/Zip LaunchPad/E- Track/Notebooks/Merged_Post_Survey_Data.csv")
```

```python
pre.columns
```

```python
# Fill NaN values in the 'Season' column with "Spring"
pre['Season'] = pre['Season'].fillna("Spring")
```

```python
pre.isnull().sum()
```

```python
pre= pre.iloc[:,:-2]
post = post.iloc[:,:-2]
pre['Year']= pd.to_datetime(pre['Timestamp']).dt.year
post['Year']= pd.to_datetime(post['Timestamp']).dt.year
```

```python
pre= pre.iloc[:,1:]
post = post.iloc[:,1:]
```

```python
pre.head()
```

```python
pre.isnull().sum()
```

```python
pre['Survey']= pre['Survey'].fillna('Pre')
post['Survey']= post['Survey'].fillna('Post')
```

```python
print(pre['Season'].unique(),post['Season'].unique())
```

```python
print(pre['Year'].unique(),post['Year'].unique())
```

### Fall 2020 Pre survey metrics. No post survey data for Fall 2020

```python
pre[(pre['Season']=='Fall')& (pre['Year']==2020)].describe().iloc[1:3]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have the mean and std values in variables
pre_mean = pre[(pre['Season']=='Fall') & (pre['Year']==2020)].describe().iloc[1]
pre_std = pre[(pre['Season']=='Fall') & (pre['Year']==2020)].describe().iloc[2]

post_mean = post[(post['Season']=='Fall') & (post['Year']==2020)].describe().iloc[1]
post_std = post[(post['Season']=='Fall') & (post['Year']==2020)].describe().iloc[2]

# Exclude the 'Year' column
pre_mean = pre_mean.drop('Year')
pre_std = pre_std.drop('Year')
post_mean = post_mean.drop('Year')
post_std = post_std.drop('Year')

# Creating a DataFrame for easy plotting
comparison_df = pd.DataFrame({
    'Pre Mean': pre_mean,
    'Pre Std': pre_std,
    'Post Mean': post_mean,
    'Post Std': post_std
})

# Plotting with error bars
fig, ax = plt.subplots(figsize=(12, 6))
categories = comparison_df.index

# Plotting pre data
ax.bar(categories, comparison_df['Pre Mean'], yerr=comparison_df['Pre Std'], capsize=5, label='Pre Fall 2020', alpha=0.7)

# Plotting post data
ax.bar(categories, comparison_df['Post Mean'], yerr=comparison_df['Post Std'], capsize=5, label='Post Fall 2020', alpha=0.7)

ax.set_title('Comparison of Mean and Standard Deviation: Pre vs Post (Fall 2020)')
ax.set_ylabel('Values')
ax.set_xticklabels(categories, rotation=45, ha='right')
ax.legend()

plt.show()
```
