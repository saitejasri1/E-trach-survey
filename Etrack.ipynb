```python
import pandas as pd

def season_and_year(df, df1, columns, years, seasons,output_file):
    # Initialize an empty list to store results
    results = []

    for year in years:
        for season in seasons:
            for column in columns:
                # Calculate mean and standard deviation for pre-survey data
                mean_df = df[(df['Year'] == year) & (df['Season'] == season)][column].describe()['mean']
                std_df = df[(df['Year'] == year) & (df['Season'] == season)][column].describe()['std']
                
                # Calculate mean and standard deviation for post-survey data
                mean_df1 = df1[(df1['Year'] == year) & (df1['Season'] == season)][column].describe()['mean']
                std_df1 = df1[(df1['Year'] == year) & (df1['Season'] == season)][column].describe()['std']
                
                # Append results to the list
                results.append({
                    'Year': year,
                    'Season': season,
                    'Column': column,
                    'Pre-Survey Mean': mean_df,
                    'Pre-Survey Std Dev': std_df,
                    'Post-Survey Mean': mean_df1,
                    'Post-Survey Std Dev': std_df1
                })

    # Convert the results list to a DataFrame
    results_df = pd.DataFrame(results)
    
    # Pivot the DataFrame to have columns for each statistic
    pivot_table = results_df.pivot_table(
        index=['Year', 'Season', 'Column'],
        values=['Pre-Survey Mean', 'Pre-Survey Std Dev', 'Post-Survey Mean', 'Post-Survey Std Dev'],
        aggfunc='first'
    ).reset_index()
    
    # Print the pivot table
    print(pivot_table.to_string(index=False))
    pivot_table.to_csv(output_file, index=False)
    print(f"Pivot table saved to {output_file}")

# Define the columns, years, and seasons


# Call the function
season_and_year(df, df1, a, Year, Season,"Season_year_pre_post.csv")
```
